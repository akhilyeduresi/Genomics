```python
#Data loading
import pandas as pd

#Data visualizations
import matplotlib.pyplot as plt
import seaborn as sns

from numpy import mean
```

```python
pd.read_excel('OmicsData_total.xlsx')
```

```python
dfsorted1=pd.read_excel('OmicsData_total.xlsx')
dfsorted1.head(5)
```

```python
#figure sizes
fig, ax = plt.subplots()
fig.set_size_inches(10, 7)

#Data subsetting
data=dfsorted1[(dfsorted1.Feature == 'RNF43') | (dfsorted1.Feature == 'RPA1')| (dfsorted1.Feature == 'ALOX15B')| (dfsorted1.Feature == 'ALOX12B')| (dfsorted1.Feature == 'STK11')| (dfsorted1.Feature == 'SOX9')| (dfsorted1.Feature == 'MCL1')| (dfsorted1.Feature == 'TSC2')| (dfsorted1.Feature == 'SEMA3A')| (dfsorted1.Feature == 'IGF2')]

#Swarnplot
#https://seaborn.pydata.org/generated/seaborn.swarmplot.html
ax = sns.swarmplot(x="Feature", y="Score", data=data, palette="Set2",hue='pSTAT3_705')

#pointplot
#https://seaborn.pydata.org/generated/seaborn.pointplot.html
sns.pointplot(x="Feature", y="Score", data=data,hue='pSTAT3_705',dodge=True, markers=["_","_"],
                   linestyles=["--", "--"],join=False,estimator=mean,color="#bb3f3f",scale=3.5)

#plt.savefig('Top 10 of mean 1.svg')
```
